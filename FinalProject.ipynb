```python
def get_html(url, params = '' ):
    r = requests.get(url, params = params)
    return r
```

```python
def get_content(html):
    soup = BeautifulSoup(html, 'html.parser')
    items = soup.find_all('div', class_ = 'review-wrap')
    reviews = []
    for item in items:
        try:
            reviews.append(
            {
                'place_name': item.find('div',class_ = 'place-name-wrap').find('a').get_text(strip = True),
                'place_rating': item.find('div',class_ = 'place-rating').get_text(),
                'review_text': item.find('div',class_ = 'review-text-wrap').get_text(strip = True),
                'link': HOST + item.find('div', class_ ='place-name-wrap').find('a').get('href')
            }
        )
        except:
            continue 
    return reviews
```

```python
def save_doc(items, path):
    with open(path, 'w', newline = '') as file:
        writer = csv.writer(file, delimiter = ',')
        writer.writerow(['name','rating','text','link'])
        for item in items:
            writer.writerow([item['place_name'],item['place_rating'],item['review_text'],item['link']])
```

```python
def parser():
    PAGENATION = int(input('Укажите количество страниц для парсинга '))
    html = get_html(URL) 
    if html.status_code == 200:
        reviews = []
        for page in tqdm_notebook(range(1, PAGENATION+1)):
            html = get_html(URL, params = {'PAGEN_1': page})
            reviews.extend(get_content(html.text))
        save_doc(reviews,CSV)
    else:
        print('Ошибка')  
```

```python
def remove_postprocess_garbage(documents):
    
    # Убираем вставки в конце
    end_garbage = 'СкрытьВсе отзывы заведенияОтветить'
    documents = documents.apply(lambda text: text.replace(end_garbage, ''))
    end_garbage = r'Скрыть[0-9][0-9]Все отзывы заведенияОтветить'
    documents = documents.apply(lambda text: re.sub(end_garbage, '', text))
    end_garbage = r'Скрыть[0-9]-[0-9]Все отзывы заведенияОтветить'
    documents = documents.apply(lambda text: re.sub(end_garbage, '', text))
    
    
    # убираем начальную неполную копию отзыва
    documents = documents.apply(lambda text: text.split('...ещё')[-1])
    
    return documents
```

### В конце люди также описывают отдельно достоинства и недостатки ресторана. Выделим каждый из них в отдельный признак

```python
def extract_marks(documents):
    good_split = 'Достоинства:'
    bad_split = 'Недостатки:'
    
    def get_good_marks(text):
        try:
            mark = text.split(good_split)[1]
            if bad_split in mark:
                mark = mark.split(bad_split)[0]
            return mark
        except:
            return 'missing'
        
    def get_bad_marks(text):
        try:
            mark = text.split(good_split)[1]
            if bad_split in mark:
                mark = mark.split(bad_split)[1]
                return mark
            return 'missing'
        except:
            try:
                mark = text.split(bad_split)[1]
                return mark
            except:
                return 'missing'
    
    def remove_marks(text):
        if good_split in text:
            return text.split(good_split)[0]
        elif bad_split in text:
            return text.split(bad_split)[0]
        return text
        
    good_marks = [get_good_marks(text) for text in documents]
    bad_marks = [get_bad_marks(text) for text in documents]
    
    documents = [remove_marks(text) for text in documents]
    
    return documents, good_marks, bad_marks
```

```python
data['text'], data['Достоинства'], data['Недостатки'] = extract_marks(data.text.copy())
```

## Приведение текстов к удобному для анализа и построения модели виду: Парсинг,  лемматизация, стемминг, удаление стоп слов

```python
nltk.download("stopwords")
ru_stopwords = stopwords.words("russian")
```

```python
def standard_preprocessing(documents, remove_stopwords=False, stopwords=None,
                           apply_stemming=False, apply_lemmatization=False):
   
    # переводим буквы в нижний регистр
    documents = documents.str.lower()

    # удаляем пунктуацию
    signs = string.punctuation + '\n\xa0«»\t—…'+'0123456789'
    documents = ["".join([' ' if symbol in signs else symbol for symbol in text]) for text in documents]
    documents = list(map(lambda text: " ".join(text.split()), documents))
   
    
    if apply_stemming == True:
        stemmer = nltk.stem.SnowballStemmer('russian')
        
        stemmer_cache = {}
        
        def stem(word):
            try:
                return stemmer_cache[word]
            except:
                stemmer_cache[word] = stemmer.stem(word)
                return stemmer_cache[word]
        
        documents = [" ".join([stem(word) for word in text.split()]) for text in documents]
    
    if apply_lemmatization == True:
        lemmatizer = pymorphy2.MorphAnalyzer()
        
        lemmatizer_cache = {}
        
        def lemmatize(word):
            if lemmatizer.word_is_known(word):
                try:
                    return lemmatizer_cache[word]
                except:
                    lemmatizer_cache[word] = lemmatizer.parse(word)[0].normal_form
                    return lemmatizer_cache[word]
            return word
        documents = [" ".join([lemmatize(word) for word in text.split()]) for text in documents] 
    
    # удаление стоп-слов
    if remove_stopwords == True:
        documents = [" ".join([word for word in text.split() if word not in stopwords]) for text in documents]
    
    # Помогает избавиться от двойных, тройных и более пробелов. В пути появится не должны были, но на всякий
    documents = list(map(lambda text: " ".join(text.split()), documents))
    
    return documents
```

```python
data['text'] = standard_preprocessing(data.text.copy(), remove_stopwords=True,
                                      stopwords=ru_stopwords, apply_lemmatization=True)
data['Достоинства'] = standard_preprocessing(data['Достоинства'].copy(), remove_stopwords=True,
                                             stopwords=ru_stopwords, apply_lemmatization=True)
data['Недостатки'] = standard_preprocessing(data['Недостатки'].copy(), remove_stopwords=True,
                                            stopwords=ru_stopwords, apply_lemmatization=True)
```

Отобразим облака слов для положительных и отрицательных отзывов

```python
def get_unigrams_frequency(good_documents, bad_documents, num_words=50):
    
    good_vocabulary = defaultdict(int)
    bad_vocabulary = defaultdict(int)
    
    def get_vocabulary(documents, vocabulary):
        for text in documents:
            tokens = text.split(' ')
            for word in tokens:
                vocabulary[word] += 1
        return vocabulary
    
    good_vocabulary = get_vocabulary(good_documents, good_vocabulary)
    bad_vocabulary = get_vocabulary(bad_documents, bad_vocabulary)
    
    good_vocabulary_df = pd.DataFrame(sorted(good_vocabulary.items(), key=lambda x: x[1], reverse=True),
                           columns=['name', 'occurence'])
    bad_vocabulary_df = pd.DataFrame(sorted(bad_vocabulary.items(), key=lambda x: x[1], reverse=True),
                           columns=['name', 'occurence'])
    
    fig, axes = plt.subplots(nrows=1, ncols=2)
    fig.set_size_inches(10, 15)
    fig.tight_layout(w_pad=10)
    
    sns.barplot(x='occurence', y='name', data=good_vocabulary_df.loc[:num_words, :], ax=axes[0], color='green')
    axes[0].set_title(f'Top {str(num_words)} words in good reviews')
    
    sns.barplot(x='occurence', y='name', data=bad_vocabulary_df.loc[:num_words, :], ax=axes[1], color='red')
    axes[1].set_title(f'Top {str(num_words)} words in bad reviews')
    
    return good_vocabulary, bad_vocabulary
```

```python
good_vocab, bad_vocab = get_unigrams_frequency(data[data['rating'] > 3.5].text.copy(),
                                               data[data['rating'] <= 3.5].text.copy(),
                                               num_words=15)
```

```python
def collocations(documents):
    cooccurence = Counter()
    occurence = Counter()

    for text in documents:
        tokens = text.split(' ')
        for i in range(len(tokens) - 1):
            cooccurence[(tokens[i], tokens[i + 1])] += 1
    
    for text in documents:
        tokens = text.split(' ')
        for word in tokens:
            occurence[word] += 1
    
    N = sum(occurence.values())
    pmi = Counter()
    min_cooccur = 20

    for pair in cooccurence:
        if cooccurence[pair] >= min_cooccur:
            pmi[pair] = np.log(cooccurence[pair] * N / occurence[pair[0]] / occurence[pair[1]])
    
    return pmi
```

### Нейронные сети

#### Transformer

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
class MultiHeadSelfAttention(layers.Layer):
    def __init__(self, embed_dim, num_heads):
        super(MultiHeadSelfAttention, self).__init__()
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        if embed_dim % num_heads != 0:
              raise ValueError(
                f"Размерность, в которую матрицы WQ, WK и WV переводят входные токены должна делиться на {str(num_heads)} - кол-во голов."
              )
        self.projection_dim = embed_dim // num_heads
        self.query_dense = layers.Dense(embed_dim)
        self.key_dense = layers.Dense(embed_dim)
        self.value_dense = layers.Dense(embed_dim)
        self.objective_dense = layers.Dense(embed_dim)  
        
    def attention(self, query, key, value):
      # query, key, value: (batch_size, num_heads, seq_len, projection_dim)
        score = tf.matmul(query, key, transpose_b=True) # (batch_size, num_heads, seq_len, seq_len)
        scale_factor = tf.math.sqrt(tf.cast(key.shape[-1], dtype=tf.float32))
        scaled_score = score / scale_factor
        weights = tf.nn.softmax(scaled_score, axis=-1)
        output = tf.matmul(weights, value) # (batch_size, num_heads, seq_len, projection_dim)
        return output, weights
    
    def separate_heads(self, x, batch_size):
      # x: (batch_size, seq_len, embed_dim) -> output: (batch_size, num_heads, seq_len, projection_dim)
        output = tf.reshape(x, (batch_size, -1, self.num_heads, self.projection_dim))
        return tf.transpose(output, perm=[0, 2, 1, 3])

    def concat_heads(self, x, batch_size):
      # x: (batch_size, num_heads, seq_len, projection_dim) -> output: (batch_size, seq_len, embed_dim)
        output = tf.transpose(x, perm=[0, 2, 1, 3])
        return tf.reshape(output, (batch_size, -1, self.embed_dim))

    def call(self, inputs):
        # inputs: (batch_size, seq_len, embedding_dim)
        batch_size = tf.shape(inputs)[0]
        query = self.query_dense(inputs) # (batch_size, seq_len, embed_dim)
        key = self.key_dense(inputs) # (batch_size, seq_len, embed_dim)
        value = self.value_dense(inputs) # (batch_size, seq_len, embed_dim)
        # (batch_size, seq_len, embed_dim) -> (batch_size, num_heads, seq_len, projection_dim)
        query = self.separate_heads( 
        query, batch_size
        ) # (batch_size, num_heads, seq_len, projection_dim)
        key = self.separate_heads(
           key, batch_size
        ) # (batch_size, num_heads, seq_len, projection_dim)
        value = self.separate_heads(
          value, batch_size
        ) # (batch_size, num_heads, seq_len, projection_dim)
        attention, weights = self.attention(
           query, key, value
        ) # (batch_size, num_heads, seq_len, projection_dim)
        concat_attention = self.concat_heads(
          attention, batch_size
        ) # (batch_size, seq_len, embed_dim)
      # т.к. мы используем residual блоки, то по хорошему нам надо на выходе дать (batch_size, seq_len, embedding_dim).
      # Но в случае трансформера, embedding_dim = embed_dim, поэтому матрица objective_dense: (embed_dim, embedding_dim=embed_dim)
        output = self.objective_dense(concat_attention) # (batch_size, seq_len, embed_dim)
        return output
```

```python
class TransformerBlock(layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, dropout_rate=0.1):
        super(TransformerBlock, self).__init__()
        self.attention_block = MultiHeadSelfAttention(embed_dim, num_heads)
        self.feedforward_block = keras.Sequential(
          [
            layers.Dense(ff_dim, activation='relu'),
            layers.Dense(embed_dim),
          ]
      )
        self.dropout1 = layers.Dropout(dropout_rate)
        self.dropout2 = layers.Dropout(dropout_rate)
        self.layernorm1 = layers.LayerNormalization(epsilon=1e-6)
        self.layernorm2 = layers.LayerNormalization(epsilon=1e-6)

    def call(self, inputs, training):
        attention_output = self.attention_block(inputs)
        attention_output = self.dropout1(attention_output, training=training)
        attention_output = self.layernorm1(attention_output + inputs)
        feedforward_output = self.feedforward_block(attention_output)
        feedforward_output = self.dropout2(feedforward_output, training=training)
        output = self.layernorm2(feedforward_output + attention_output) # (batch_size, seq_len, embed_dim)
        return output
```

```python
class TokenAndPositionEmbedding(layers.Layer):
    def __init__(self, maxlen, vocab_size, embed_dim):
        super(TokenAndPositionEmbedding, self).__init__()
        self.token_emb = layers.Embedding(input_dim=vocab_size, output_dim=embed_dim, mask_zero=True)
        self.pos_emb = layers.Embedding(input_dim=maxlen, output_dim=embed_dim)

    def call(self, x):
      # x : (batch_size, seq_len, maxlen) -> (batch_size, seq_len, embed_dim)
        maxlen = x.shape[-1]
        x = self.token_emb(x)
        positions = tf.range(start=0, limit=maxlen, delta=1)
        positions = self.pos_emb(positions) 
        return x + positions
```

#### Посмотрим на Attention головы

    Функция принимает на вход предложение и номер слова в этом предложении, а также N - кол-во топ слов для этого слова.
    Функция находит топ N слов для выбранного слова на основе attention block'а

```python
def get_top_n_words(sentence, word_idx=0, n=5):

    inp = model.input
    attn_weights = model.layers[2].trainable_weights

    query_weights, query_bias = attn_weights[0], attn_weights[1]
    key_weights, key_bias = attn_weights[2], attn_weights[3]

    embedding = keras.Model(inputs=model.input, outputs=model.layers[1].output)
    ex_embed = tf.transpose(embedding.predict(sentence), perm=[1, 0, 2]) # (batch_size=1, seq_len=300, embed_dim=32)

    def attention(query, key):
      # query, key, value: (batch_size, num_heads, seq_len, projection_dim)
        score = tf.matmul(query, key, transpose_b=True) # (batch_size, num_heads, seq_len, seq_len)
        scale_factor = tf.math.sqrt(tf.cast(key.shape[-1], dtype=tf.float32))
        scaled_score = score / scale_factor
        weights = tf.nn.softmax(scaled_score, axis=-1)
        return weights

    query = tf.math.add(tf.matmul(ex_embed, query_weights), query_bias)
    key = tf.math.add(tf.matmul(ex_embed, key_weights), key_bias)

    ex_weights = attention(query, key).numpy()

    top_words = dict(sorted(dict(zip(sentence, ex_weights[0, word_idx])).items(), key=lambda x: x[1], reverse=True)[:n])

    cur_word = token2word.get(sentence[word_idx])

    words = []
    for word in top_words.keys():
        words.append(token2word.get(word))

    text_sentence = []
    for token in sentence:
        if token == 0:
            break
        text_sentence.append(token2word.get(token))
    text_sentence = " ".join(text_sentence)

    print("Все предложение:\n", text_sentence)

    print(f"Для слова \"{cur_word}\" самые важные слова:\n", words)

    return cur_word, words
```
